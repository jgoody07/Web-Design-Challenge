```python
#Looking at Price vs. Weight
plt.scatter(brooks["Weight(g)"], 
            brooks["Midsole Drop(mm)"], c=brooks["Price"],
            edgecolor="black", linewidths=1, marker="o", 
            cmap='plasma', alpha=1)

plt.title('Shoe Weight vs. Midsole Drop(mm)')
plt.ylabel("Midsole Drop(mm)")
plt.xlabel("Weight (g)")
```

```python
boston = pd.read_csv('Resources/Female_Elite_Boston.csv')
boston
```

```python
import datetime as dt
times = []
def second_counter(series):
    for row in series:
        h,m,s = row.split(':')
        times.append(int(dt.timedelta(hours=int(h),minutes=int(m),seconds=int(s)).total_seconds()))
       
# boston['Time'].apply(second_counter())
```

```python
second_counter(boston["Time"])
```

```python
times
```

```python
boston["Seconds"] = pd.Series(times)
boston
```

```python
boston.Year = boston.Year.astype('str')
```

```python
boston.Year = boston.Year.str.lstrip('\n           ')
boston.Year = boston.Year.str.lstrip('\n')
```

```python
boston.plot(x=boston.Year, y=boston.Seconds)
```
