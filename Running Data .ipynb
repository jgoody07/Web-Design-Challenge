```python
import datetime as dt
times = []
def second_counter(series):
    for row in series:
        h,m,s = row.split(':')
        times.append(int(dt.timedelta(hours=int(h),minutes=int(m),seconds=int(s)).total_seconds()))
       
# boston['Time'].apply(second_counter())
```

```python
second_counter(boston["Time"])
```

```python
times
```

```python
boston["Seconds"] = pd.Series(times)
boston
```

```python
boston.Year = boston.Year.astype('str')
```

```python
boston.Year = boston.Year.str.lstrip('\n           ')
boston.Year = boston.Year.str.lstrip('\n')
```
